```python
def is_trivial(groups, seats):
    return (sum(groups) + len(groups) - 1) <= seats

def is_feasible(groups, seats):
    return sum(groups) - len(groups) <= seats - 1
```

```python
def output_seating_chart(groups):
    seat_num = 1
    for people in range(len(groups)):
        for person in range(groups[people]):
            print("Person " + str(person + 1) + " in Group " + str(people + 1) + " sits in seat " + str(seat_num))
            seat_num = seat_num + 1
        print("SOCIAL DISTANCING in seat " + str(seat_num))
        seat_num = seat_num + 1
```

```python
def find_solution(accepted, rejected, groups, seats):
    
    print("NEW ITERATION")
    
    # find the largest group in rejected and add to accepted
    max_index = max(rejected, key=rejected.get)
    
    if(groups[max_index] <= seats): # ensure the value being added won't invalidate the solution 
        accepted[max_index] = rejected.pop(max_index)
        print("group " + str(max_index) + "of size" + str(groups[max_index]) + "has been added")
    
    else:
        rejected.pop(max_index)
        print("group " + str(max_index) + "of size" + str(groups[max_index]) + "has been added PREMANATLY")
    
    # remove smallest values until feasible
    while(not is_trivial(accepted.values(), seats)):
        min_index = min(accepted, key = accepted.get)
        rejected[min_index] = accepted.pop(min_index) 
        
        print("group " + str(min_index) + "of size" + str(groups[min_index]) + "has been removed")

    # add back next smallest rejected value until just under the maximum
    can_be_maximized = True
    while(can_be_maximized):
        min_index = min(rejected, key=rejected.get)
        
        if(sum(accepted.values()) + groups[min_index] > seats):
            can_be_maximized = False
        else: 
            accepted[min_index] = rejected.pop(min_index)
            print("group " + str(min_index) + "of size" + str(groups[min_index]) + "has been added")


            
    
    return accepted 
```

```python
def find_best_solution(groups,seat):
    # account for the buffer by adding 
    for index in range(len(groups)):
        groups[index] = groups[index] + 1   
    seat = seat + 1 #add 1 to account for the excess seat allocated by the inflation above
    
    #initialize 
    accepted = dict()
    rejected = dict()
    for group, size in enumerate(groups):
        if(sum(accepted.values()) + size <= seat):
            accepted[group] = size
        else:
            rejected[group] = size
            
    incumbent = accepted.copy()
    print("initial incumbent is: " + str(incumbent))
    
    unchanged = 0
    while(unchanged <= 10):
        accepted, rejected = find_solution(accepted.copy(), rejected.copy(), groups, seat)
        
        if(sum(accepted.values())-len(accepted.keys()) > sum(incumbent.values()) - len(incumbent.keys())):
            unchanged = 0;            
            print()
            print("OPTIMIZATION")
            print("the old incumbent is" + str(incumbent))
            print("the new incumbent is" + str(accepted))
            print(str(sum(incumbent.values())) + "improved to" + str(sum(accepted.values())))
            print()
            
            incumbent = accepted.copy()
        else:
            unchanged = unchanged + 1
            print()
            print("inciumbent did not change")
            print(str(sum(incumbent.values())) + "is better than" + str(sum(accepted.values())))
            print("The imcumbent looks like: " + str(incumbent))
            print("We will iterate on " + str(accepted))
            print()
    return incumbent
```

```python
def find_optimal_seating(groups, seats):
    if(is_trivial(groups, seats)):
        output_seating_chart(groups)
    else:
        accepted_groups = find_best_solution(groups, seats)
        seat_num = 1
        for group in accepted_groups.keys():
            for person in range(accepted_groups[group] - 1):
                print("Person " + str(person + 1) + " in Group " + str(group + 1) + " sits in seat " + str(seat_num))
                seat_num = seat_num + 1
            print("SOCIAL DISTANCING in seat " + str(seat_num))
            seat_num = seat_num + 1
        print("total seats allocated is " + str(sum(accepted_groups.values()) - len(accepted_groups.keys())))
        print("capacity rate = " + str((sum(accepted_groups.values()) - len(accepted_groups.keys())) * 100 / seats ) + "%")
```

```python
groups = [10,3,6,4,76,5,19,19,19]
seats = 60
find_optimal_seating(groups, seats)
```
